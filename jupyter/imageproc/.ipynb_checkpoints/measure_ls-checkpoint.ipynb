```python
img = cv2.imread('../../image/sample_ls.png', cv2.IMREAD_ANYDEPTH)
diff = np.diff(img[0].astype(np.int16))
indices = np.where(np.abs(diff) > 40)[0] + 1
# indices = np.where(np.abs(diff) > 40)
print(indices)
fig, ax = plt.subplots(2, 1, figsize=(17, 8))
ax[0].plot([i for i in range(len(img[0]))], img[0])
ax[1].plot([i for i in range(1,len(img[0]-1))], diff)
for i in indices:
    ax[0].axvline(x=i, color='red')
    ax[1].axvline(x=i, color='red')

# Sobelフィルタを用いて勾配を計算
gradient_x = cv2.Sobel(img[0], cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(img[0], cv2.CV_64F, 0, 1, ksize=3)

# 勾配の大きさを計算
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# 勾配が最大となる位置を求める
max_gradient_position = np.unravel_index(np.argmax(gradient_magnitude), gradient_magnitude.shape)

print("Max Gradient Position (row, column):", max_gradient_position)
fig, ax = plt.subplots()
ax.imshow(img, cmap='gray')
ax.axvline(x=max_gradient_position[0])
```

```python
# Cannyエッジ検出を適用
edges = cv2.Canny(img, threshold1=100, threshold2=200)  # 適切なしきい値を設定

# エッジを描画して表示
plt.figure(figsize=(8, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image, ImageDraw

# 画像のサイズと背景色を設定
width, height = 512, 512
background_color = (0, 0, 0)  # 黒

# 画像を作成
image = Image.new('RGB', (width, height), background_color)
draw = ImageDraw.Draw(image)

# コンタクトホールのパラメータを設定
num_contact_holes = 80  # コンタクトホールの数
contact_hole_radius = 5  # コンタクトホールの半径
spacing = 10  # コンタクトホールの間隔

# コンタクトホールを描画
for y in range(contact_hole_radius, height, spacing):
    for x in range(contact_hole_radius, width, spacing):
        for dy in range(-contact_hole_radius, contact_hole_radius + 1):
            for dx in range(-contact_hole_radius, contact_hole_radius + 1):
                distance = np.sqrt(dx ** 2 + dy ** 2)
                if distance <= contact_hole_radius:
                    intensity = int(255 * (1 - distance / contact_hole_radius))
                    noise = np.random.randint(-20, 20)  # ランダムなノイズを生成
                    intensity += noise
                    intensity = max(0, min(255, intensity))  # 輝度を0から255の範囲に制限
                    draw.point((x + dx, y + dy), fill=(intensity, intensity, intensity))

# 画像を保存
image.save('CH_sample_with_noise.png')

print("画像が作成されました。")
```

```python
# Nearest neighbor
def nearest_neighbor(img, x, y):
    x_ = int(x+0.5)
    y_ = int(y+0.5)
    return img[y_][x_]

# Biliner
# def biliner(img, x, y):
#     intensity = 

# Bicubic

def estimate_dimensions(image_path, pixel_size):
    # Load the image
    image = cv2.imread(image_path, 0)  # Read the image as grayscale

    # Apply edge detection
    edges = cv2.Canny(image, 100, 200)  # Adjust the threshold values as needed

    # Find contours
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Find the longest contour (assuming it represents the vertical line)
    longest_contour = max(contours, key=cv2.contourArea)

    # Calculate the length of the contour in pixels
    contour_length = cv2.arcLength(longest_contour, True)

    # Calculate the estimated length in the real world based on the pixel size
    estimated_length = contour_length * pixel_size

    return estimated_length
```
