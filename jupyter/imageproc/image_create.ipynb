```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
```

## Create sample images

```python
image_size = (512, 512)
num_images = 13
brightness = 23040
noise_level = 1000

for i in range(num_images):
    _img = np.ones(image_size, dtype=np.uint16) * brightness
    _noise = np.random.normal(0, noise_level, image_size).astype(np.uint16)
    _noisy_img = np.clip(_img + _noise, 0, 65535)
    _dir = "./image/"
    _filename = f"img{i + 1}.png"
    cv2.imwrite(_dir + _filename, _noisy_img.astype(np.uint16))
    # 
    plt.subplot(3, 5, i + 1)
    plt.imshow(_noisy_img, cmap='gray')
    plt.title(f'Image {i + 1}')
    plt.axis('off')

plt.show()
```
