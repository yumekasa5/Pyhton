```python
import sys


import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

from pathlib import Path
cur_dir = Path().resolve()
sys.path.append(str(cur_dir) + "/module")
```

## Poisson分布

```python
# 試行回数
n = 1000
# 確率
p = 0.002
# Poisson分布の平均値
lambda_ = n * p

# Poisson分布の生成
poisson_dist = poisson(mu=lambda_)

# サンプルデータの生成
sample_data = poisson_dist.rvs(size=n)

# ヒストグラムの表示
plt.hist(sample_data, bins=20, density=True, alpha=0.7, rwidth=0.8)

# 理論的な確率質量関数のプロット
x = np.arange(0, np.max(sample_data) + 1)
pmf = poisson_dist.pmf(x)
plt.plot(x, pmf, label='Poisson PMF')

# グラフの設定
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Poisson Distribution (lambda={})'.format(lambda_))
plt.legend()

# グラフの表示
plt.show()
```

```python
# 確率変数X=0のときの確率を求める
p0 = poisson_dist.pmf(0)
# 確率変数X=1のときの確率を求める
p1 = poisson_dist.pmf(1)
# 確率変数X=2のときの確率を求める
p2 = poisson_dist.pmf(2)

# 結果の表示
print(p0)
print(p1)
print(p2)
```
