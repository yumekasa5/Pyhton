# QR Code Read
* OpenCv QRCodeDetectorを使う
* detectAndDecodeMulti()メソッドの返り値は以下の通り
    * bool retval : 検出結果
    * Tuple<string> detect_info : 文字列のタプル
    * List<points> : QRCodeの四隅の点の座標 

```python
img = cv2.imread(r'../../QR/qr1.jpg', cv2.IMREAD_ANYDEPTH)
img2 = cv2.imread(r'../../QR/qr2.jpg', cv2.IMREAD_ANYDEPTH)
img3 = cv2.imread(r'../../QR/qr3.jpg', cv2.IMREAD_ANYDEPTH)

#detect QR code
qcd = cv2.QRCodeDetector()
retval, detect_info, points, straight_qrcode = qcd.detectAndDecodeMulti(img)
retval2, detect_info2, points2, straight_qrcode2 = qcd.detectAndDecodeMulti(img2)
retval3, detect_info3, points3, straight_qrcode3 = qcd.detectAndDecodeMulti(img3)

#result
image_list = [img, img2, img3]
retval_list = [retval, retval2, retval3]
detect_info_list = [detect_info, detect_info2, detect_info3]
points_list = [points, points2, points3]

fig, ax = plt.subplots(3, 1, figsize=(10, 20))
for a, qr_image, ret, info, point in zip(ax, image_list, retval_list, detect_info_list, points_list):
    a.imshow(qr_image, cmap='gray')
    if(ret):
        a.scatter(point[0][0][0], point[0][0][1], s = 50, marker='o', color='r')
        a.scatter(point[0][1][0], point[0][1][1], s = 50, marker='o', color='r')
        a.scatter(point[0][2][0], point[0][2][1], s = 50, marker='o', color='r')
        a.scatter(point[0][3][0], point[0][3][1], s = 50, marker='o', color='r')    
```
