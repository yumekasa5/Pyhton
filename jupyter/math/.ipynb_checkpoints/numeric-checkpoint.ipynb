```python
import glob
import math
import os
import errno
import sys

import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import skimage.measure as ski
import re

from pathlib import Path
cur_dir = Path().resolve()
sys.path.append(str(cur_dir) + "/module")
```

```python
# オイラー法を用いた数値積分
x0 = 0  # 初期位置[m]
v0 = 0  # 初速度[m/s]
M = 1000 # 質量[kg]
F = 500  # 推進力[N]
alpha = F/M # 加速度[m/s^2]
t_end = 100 # シミュレーション終了時刻[s]
fig, ax = plt.subplots()

# 厳密解
delta = 1 # 時間刻み[s]
datamax = int(t_end/delta)
time = np.zeros(datamax+1)
x_true = np.zeros(datamax+1)
for i in range(1, len(time)):
    time[i] = i * delta
    x_true[i] = 0.5 * alpha * time[i]**2 + v0 *time[i] + x0
ax.plot(time, x_true, label='True')

# オイラー法
delta = 1
datamax = int(t_end/delta)
time1 = np.zeros(datamax+1)
v_euler1 = np.zeros(datamax+1)
x_euler1 = np.zeros(datamax+1)

for i in range(0, datamax):
    time1[i+1] = time[i] + delta
    v_euler1[i+1] = v_euler1[i] + alpha * delta
    x_euler1[i+1] = x_euler1[i] + v_euler1[i] * delta
ax.plot(time1, x_euler1, label='Euler_1sec')
ax.legend()
ax.set_xlabel('Time[s]')
ax.set_ylabel('Position[m]')
plt.show()

# ルンゲ・クッタ法
```
