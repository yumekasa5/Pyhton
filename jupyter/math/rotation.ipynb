```python
import numpy as np
import matplotlib.pyplot as plt

p0 = np.array([1, 1])
p1 = np.array([1, 0])
vec = p1 - p0
inner = np.inner(p0, p1)
l_p0 = np.linalg.norm(p0)
l_p1 = np.linalg.norm(p1)
theta = np.arccos(inner / (l_p0 * l_p1))
theta = np.rad2deg(theta)
print(theta)

fig, ax = plt.subplots()
ax.set_xlim([-1, 5])
ax.set_ylim([0, 5])
ax.quiver(0, 0, 3,  3, color='red', angles='xy', scale_units='xy', scale=1)
ax.quiver(0, 0, 0,  3, color='blue', angles='xy', scale_units='xy', scale=1)
plt.show()
```
