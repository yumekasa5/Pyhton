```python
import numpy as np
import matplotlib.pyplot as plt

# データを生成
x = np.linspace(0, 10, 100)
y = np.sin(x)

# FigureとAxesを描画
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label="sin(x)")
ax.grid()

# 注釈と矢印を入れる位置
x_position = 4
y_position = np.sin(x_position)

# 注釈を入れる
ax.annotate("point A", xy=(x_position, y_position), xytext=(x_position, 1),
            arrowprops=dict(arrowstyle="->", color="red"))

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sinusoidal Function with Annotation")
plt.legend()
plt.show()
```
