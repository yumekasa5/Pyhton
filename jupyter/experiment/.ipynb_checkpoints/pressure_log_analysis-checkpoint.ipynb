```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
import sys
import glob
```

```python
path = r"C:\work\python\jupyter\experiment\data\csv\pressure.csv"
df = pd.read_csv(path)
df["DateTime"] = pd.to_datetime(df["Date"] + " " + df["Time"])
df.set_index("DateTime")
# df = df.loc["2024-07-01 00:00:00" : "2024-07-01 02:00:00"]
df.head()
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("Time")
ax.set_ylabel("Pressure(Pa)")
ax.plot(df.index, df["A"], label="A")
ax.plot(df.index, df["B"], label="B")
ax.plot(df.index, df["C"], label="C")

ax.legend()

formatter = mdates.DateFormatter("%H:%M:%S")
ax.xaxis.set_major_formatter(formatter)
plt.gcf().autofmt_xdate()  # 目盛りの自動回転配置
```
